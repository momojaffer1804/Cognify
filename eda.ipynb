```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figsize=(10,6)
sns.heatmap(df.corr(),  cmap='RdYlGn_r')
plt.show()
```

- The heatmap shows **negative correlation** (dark blue color) between **Diagnosis** and some brain-related features.
- These features include **MMSE**, **ADL**, **FunctionalAssessment**, **DifficultyCompletingTasks**, **Disorientation**, and **MemoryComplaints**.
- Negative correlation means that when these scores **increase**, the chance of being diagnosed with Alzheimer’s **decreases**.
- In simple terms, **better memory, thinking ability, and daily functioning** are linked to a **lower risk of Alzheimer’s**.
- This result makes sense medically, because Alzheimer’s disease mainly affects memory and daily activities.
- If these features had shown a **positive correlation**, it would have been unusual and could indicate problems with the data.

```python
boxplot_features = [
    "MMSE",
    "ADL",
    "FunctionalAssessment"
]


countplot_features = [
    "Disorientation",
    "MemoryComplaints",
    "DifficultyCompletingTasks"
]
```

```python
fig, axes = plt.subplots(3, 2, figsize=(6,8))


for i, feature in enumerate(boxplot_features):
    sns.boxplot(
        x="Diagnosis",
        y=feature,
        data=df,
        ax=axes[i, 0]
    )
    axes[i, 0].set_title(f"{feature} vs Diagnosis")
plt.tight_layout()
plt.show()
```

- Continuous cognitive and functional features such as ADL, FunctionalAssessment, and MMSE show clear separation between diagnosed and non-diagnosed individuals.
- Patients with Alzheimer’s consistently have lower median scores, indicating reduced cognitive ability and daily functioning.
- Binary features such as Disorientation, MemoryComplaints, and DifficultyCompletingTasks show limited variation in boxplots and therefore we use countplot


```python
fig, axes = plt.subplots(3, 2, figsize=(6,8))  

for i, feature in enumerate(countplot_features):
    sns.countplot(
        x=feature,
        hue="Diagnosis",
        data=df,
        ax=axes[i, 1]
    )


plt.tight_layout()
plt.show()
```

 Binary Cognitive & Functional Indicators – Key Takeaways
- Most individuals do not exhibit disorientation, memory complaints, or difficulty completing tasks.
- When these symptoms are present, a higher proportion of individuals are diagnosed with Alzheimer’s.
- This indicates that disorientation, memory complaints, and task difficulty are associated with Alzheimer’s diagnosis.
- However, these symptoms are not observed in all diagnosed cases, particularly in early stages of the disease.
- Therefore, these features serve as supportive indicators and are most effective when combined with continuous cognitive and functional assessment scores such as MMSE, ADL, and FunctionalAssessment.

### Cluster & Separation

```python
main_features =  [
    "MMSE",
    "ADL",
    "FunctionalAssessment",
    "Disorientation",
    "MemoryComplaints",
    "DifficultyCompletingTasks"

]
sns.pairplot(df[main_features + ["Diagnosis"]], hue="Diagnosis")
plt.show()
```

Cluster & Separation Insights
- Diagonal distributions show that Alzheimer’s patients tend to have lower MMSE, ADL, and FunctionalAssessment scores.
- Off-diagonal scatter plots show significant overlap between diagnosed and non-diagnosed individuals, indicating the absence of clear linear separation.
- This suggests that no single feature or feature pair can independently classify the diagnosis.
- The observed patterns support the use of multi-feature and non-linear models for prediction.
